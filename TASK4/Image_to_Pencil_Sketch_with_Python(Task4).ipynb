## Importing Required Libraries

```python
import cv2
from google.colab.patches import cv2_imshow # since cv2.imshow is not working in COLAB, so we use this module
from PIL import Image # Reshape the given image
```

```python
image = Image.open('/content/image.jpg')
image=image.resize((300,300))
image.save('oimage.jpg')
```

## Importing the image

```python
image = cv2.imread("oimage.jpg")
cv2_imshow(image)
cv2.waitKey(0)
```

## Converting Image to Gray-scale image

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
cv2.waitKey(0)
```

## Invert the new grayscale image

```python
inverted_image = 255 - gray_image
cv2_imshow(inverted_image)
cv2.waitKey()
```

## Blur the image

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21),sigmaX=0,sigmaY=0)
```

## Inverting the Blurred Image

```python
inverted_blurred = 255 - blurred
```

## Convert the image into a pencil sketch

```python
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```

### **Original & Pencil Sketch**

```python
print("Original")
cv2_imshow(image)
print("\n")
print("Sketch")
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```
